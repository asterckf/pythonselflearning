# Histogram

```python
%matplotlib inline 
```

```python
#create a random data set 
dataset1=randn(100) 

plt.hist(dataset1,color='indianred')
```

```python
dataset1=randn(100)
dataset2=randn(50)

plt.hist(dataset1, normed=True, color='indianred' , alpha=0.5, bins=20)
plt.hist(dataset2, normed=True, alpha=0.5, bins=20)
```

# Joint distribution using joint plots 

```python
#create two random list data sets

data1=randn(3000)
data2=randn(3000)

sns.jointplot(data1,data2)
```

```python
data1=randn(1000)
data2=randn(1000)

sns.jointplot(data1,data2 , kind='hex', color='red')
```

# Kernel Density Estimation Plot

```python
#create a random dataset
dataset=randn(25)

plt.hist(dataset, alpha=0.3)
sns.rugplot(dataset, color='darkblue')
```

```python
sns.rugplot(dataset, color='blue')

for bw in np.arange(0.5,2,0.25):
   sns.kdeplot(dataset, bw=bw, lw=5, label=bw)
```

# Combining Plot Styles

```python
#create new dataset 
dataset= randn(100)

sns.distplot(dataset, bins=25, color='red')
```

```python
from pandas import Series

dataset= randn(100)
ser1=Series(dataset, name='MY_DATA')

sns.distplot(ser1,bins=25, 
             kde_kws ={'color':'red','label':'KDE PLOT'}, 
             hist_kws={'color':'green','label':'HISTOGRAM'})
```
