# Heatmaps and clustered Matrices

```python
flight_dframes=sns.load_dataset('flights')
flight_dframes.head()
```

```python
#create pivot 
flight_dframes=flight_dframes.pivot("month","year","passengers")
flight_dframes
```

```python
sns.heatmap(flight_dframes)
```

```python
sns.heatmap(flight_dframes,annot=True, fmt='d')
```

```python
sns.clustermap(flight_dframes)
```
