```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g= nx.read_edgelist('facebook_combined.txt',create_using= nx.Graph() , nodetype=int)
```

```python
g.number_of_nodes() #number of nodes
```

```python
g.number_of_edges() #number of edges
```

```python
sp = nx.spring_layout(g)
```

```python
plt.axis('off')
```

```python
nx.draw_networkx(g,pos=sp, with_labels=False,node_size=35)
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5,10.5)
fig.savefig('test2png.png', dpi=100)
plt.show()
```

```python
graph = nx.Graph()
number_of_friends = graph.degree()

for person in number_of_friends:
    print("{} had {} friends".format(person,number_of_friends(person)))
```
